# Training and Test Split of the Data

```python
X = ecomm[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
Y = ecomm['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

# Training the Model
Now its time to train the model on our training data!

#Import LinearRegression from sklearn.linear_model

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train, Y_train)
print(lm.coef_)
print(lm.intercept_)
```

```python
#pair the feature name with the coefficients

print(list(zip(X.columns,lm.coef_)))
```

# Predicting Test Data
Now that we have fit our model, let's evaluate its performance by predicting the test values

Use lm.predict() to predict the X_test set of the data.

```python
Y_pred = lm.predict(X_test)
print(Y_pred)
```

```python
lm.score(X_train,Y_train)
```

# Creating a scatterplot of the actual test values versus the predicted values.

```python
plt.scatter(Y_test,Y_pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics

print('MAE :'," ", metrics.mean_absolute_error(Y_test,Y_pred))
print('MSE :'," ", metrics.mean_squared_error(Y_test,Y_pred))
print('RMSE :'," ", np.sqrt(metrics.mean_squared_error(Y_test,Y_pred)))
```

```python
new_ecom = pd.DataFrame()
new_ecom = X_test

new_ecom['Actual'] = Y_test
new_ecom['Predicted'] = Y_pred

new_ecom
```
